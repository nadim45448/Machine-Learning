2.B: Don't drop.Now we will put some valule in the null place.For numerical null value it can be the mean/median/mode of the column and for the catagorial value it can be the most frequent value in the column

```python
df2['Age']=df2['Age'].fillna(df2['Age'].mean()) #replace by mean
#df2['Age']=df2['Age'].fillna(df2['Age'].median()) #replace by median
```

```python
most_frequent_catagory=df2['Country'].mode().iloc[0] #Find the most frequent catagory
df2['Country'].fillna(most_frequent_catagory, inplace=True) #Replacing with most frequent value
```
